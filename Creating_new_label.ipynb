```python
import pandas as pd
import numpy as np
import imutils
import glob
import cv2
import shutil
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
tqdm_notebook().pandas()
```

```python
label_df = pd.read_csv('C:/Thesis projects/Face_clusteing/img_align_celeba_1/list_attr_celeba_1.csv')
column_list = pd.Series(list(label_df.columns)[1:])
print(column_list)
def label_generator(row):
    return(' '.join(column_list[[True if i==1 else False for i in row[column_list]]]))

label_df['label'] = label_df.progress_apply(lambda x: label_generator(x), axis=1)
label_df = label_df.loc[:,['image_id','label']]
label_df.to_csv('C:/Thesis projects/Face_clusteing/img_align_celeba_1/labels.csv')
```

```python
## Attachhing label to correct file names
item_list = glob.glob('C:/Thesis projects/Face_clusteing/img_align_celeba_1/*/*.jpg')
print(item_list)
item_df = pd.DataFrame({'image_name':pd.Series(item_list).apply(lambda x: '/'.join(x.split('/')[-2]))})
item_df['image_id'] = item_df.image_name.apply(lambda x: x.split('/')[1])
```

```python
## Creating final label set
label_df = pd.read_csv('C:/Thesis projects/Face_clusteing/img_align_celeba_1/labels.csv')
label_df = label_df.merge(item_df, on='image_id', how='inner')
label_df.rename(columns={'label':'tags'}, inplace=True)
label_df.loc[:,['image_name','tags']].to_csv('C:/Thesis projects/Face_clusteing/img_align_celeba_1/labels.csv', index=False)
```
